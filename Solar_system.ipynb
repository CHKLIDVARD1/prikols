```python
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
import numpy as np
import seaborn as sns
```

```python
%matplotlib notebook
```

```python
cords = []
sns.set_style("dark")
lens = np.array([1, 2, 3, 5, 7, 10, 14, 20])
nums = np.random.uniform(0,10, 8)
for i in np.linspace(0,100,5000):
    argsx = lens * np.cos(np.pi*i + nums)
    argsy = lens * np.sin(np.pi*i + nums)
    cords.append(list(zip(argsx, argsy)))
cords = np.array(cords)
colors = sns.color_palette("tab10")

x = np.linspace(0,2,500)
fig, ax = plt.subplots()
ax.plot(0, 0, 'yo')
ax.plot(1 * np.cos(np.pi*x), 1 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(2 * np.cos(np.pi*x), 2 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(3 * np.cos(np.pi*x), 3 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(5 * np.cos(np.pi*x), 5 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(7 * np.cos(np.pi*x), 7 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(10 * np.cos(np.pi*x), 10 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(14 * np.cos(np.pi*x), 14 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)
ax.plot(20 * np.cos(np.pi*x), 20 * np.sin(np.pi*x), 'cadetblue', linewidth = 0.5)

ax.set_xlim(-21,21)
ax.set_ylim(-21,21)
graph = ax.scatter([],[], marker='o', s=15)
def animate(i):
    graph.set_offsets(cords[i])
    graph.set_facecolor(colors[:i])
    return graph
    
animation = FuncAnimation(fig, func=animate, frames = range(np.shape(cords)[0]), interval=10, repeat = True)
```
